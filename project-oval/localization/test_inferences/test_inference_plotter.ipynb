```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# Open the csv file
csv_file_path_model_1 = "RCNN_FFNN_Normal_Shallow.csv"
csv_file_path_model_2 = "YOLO_FFNN_normal_shallow.csv"

# Removing the outlier and 2112 is the total image size 
image_names = [i for i in range(1, 2113)]
with open(csv_file_path_model_1, 'r') as file:
    reader = csv.DictReader(file)
    
    processing_times_model_1 = []

    # Iterate over the rows in the CSV and extract the columns you need
    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_1.append(float(row['Processing Time (s)']) * 1000)

with open(csv_file_path_model_2, "r") as file:
    reader = csv.DictReader(file)

    processing_times_model_2 = []

    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_2.append(float(row['Processing Time']) * 1000)

# Display the first 20 values in milliseconds
print("First 20 processing times for Model 2 (in milliseconds):", processing_times_model_2[:20])   
print("Total image names:", len(image_names))
print("Total processing times for Model 1:", len(processing_times_model_1))
print("Total processing times for Model 2:", len(processing_times_model_2))
      
# Check if the lengths are consistent before plotting
if len(image_names) == len(processing_times_model_1) == len(processing_times_model_2):
    
    # Add jitter (optional) to spread points
    jitter_factor = 0.005
    processing_times_model_1_jittered = np.array(processing_times_model_1) + np.random.normal(0, jitter_factor, len(processing_times_model_1))
    processing_times_model_2_jittered = np.array(processing_times_model_2) + np.random.normal(0, jitter_factor, len(processing_times_model_2))

    # Plotting both datasets on the same axes
    fig = plt.figure(figsize=(12, 8))       
    ax1 = fig.add_subplot(111)

    # Scatter plot for Model 1 (RCNN)
    ax1.scatter(image_names[1:650], processing_times_model_1_jittered[1:650], color='blue', s=30, alpha=0.3, label="RCNN+FFNN")
    
    # Scatter plot for Model 2 (YOLO)
    ax1.scatter(image_names[1:650], processing_times_model_2_jittered[1:650], color='red', s=30, alpha=0.6, label="YOLO+FFNN")
                                        
    # Add titles and labels
    plt.title('Model Inference Time Comparison (Logarithmic Scale with Jitter)')
    plt.xlabel('Image Index')
    plt.ylabel('Processing Time (ms)')
    plt.xticks(rotation=45, ha='right')

    # Set a logarithmic scale for the y-axis
    ax1.set_yscale('log')

    # Adjust y-axis limits to zoom in on the area of interest
    ax1.set_ylim([10, 3000])  # Adjust based on your data range

    # Add a legend with larger font size
    plt.legend(fontsize=12)  # You can increase the fontsize as per your preference

    # Add a grid
    ax1.grid(True)  # You can control grid style by passing arguments like linestyle, color, etc.

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.savefig('RCNN_VS_YOLO+FFNN.jpg', dpi=300) 
    plt.show()
    

else:
    print("Lengths of the datasets do not match. Cannot plot!")
```

```python
# Open the csv file 
csv_file_path_model_1 = "RCNN_FFNN_Normal_Shallow.csv"
csv_file_path_model_2 = "YOLO_FFNN_normal_shallow.csv"
csv_file_path_model_3 = "RCNN_LSTM_Normal_Shallow.csv"
csv_file_path_model_4 = "YOLO_LSTM_normal_shallow.csv"

# Removing the outlier and 2112 is the total image size 
image_names = [i for i in range(1, 2113)]

with open(csv_file_path_model_1, 'r') as file:
    reader = csv.DictReader(file)
    processing_times_model_1 = []
    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_1.append(float(row['Processing Time (s)']) * 1000)

with open(csv_file_path_model_2, "r") as file:
    reader = csv.DictReader(file)
    processing_times_model_2 = []
    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_2.append(float(row['Processing Time']) * 1000)

with open(csv_file_path_model_3, "r") as file:
    reader = csv.DictReader(file)
    processing_times_model_3 = []
    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_3.append(float(row['Processing Time (s)']) * 1000)

with open(csv_file_path_model_4, "r") as file:
    reader = csv.DictReader(file)
    processing_times_model_4 = []
    for row in reader:
        # Convert to milliseconds by multiplying by 1000
        processing_times_model_4.append(float(row['Processing Time']) * 1000)

# Display the first 20 values in milliseconds for model 2
print("First 20 processing times for Model 2 (in milliseconds):", processing_times_model_2[:20])   
print("Total image names:", len(image_names))
print("Total processing times for Model 1:", len(processing_times_model_1))
print("Total processing times for Model 2:", len(processing_times_model_2))
      
# Check if the lengths are consistent before plotting
if len(image_names) == len(processing_times_model_1) == len(processing_times_model_2) == len(processing_times_model_3) == len(processing_times_model_4):
    
    # Add jitter (optional) to spread points
    jitter_factor = 0.005
    processing_times_model_1_jittered = np.array(processing_times_model_1) + np.random.normal(0, jitter_factor, len(processing_times_model_1))
    processing_times_model_2_jittered = np.array(processing_times_model_2) + np.random.normal(0, jitter_factor, len(processing_times_model_2))
    processing_times_model_3_jittered = np.array(processing_times_model_3) + np.random.normal(0, jitter_factor, len(processing_times_model_3))
    processing_times_model_4_jittered = np.array(processing_times_model_4) + np.random.normal(0, jitter_factor, len(processing_times_model_4))
    
    # Plotting all datasets on the same axes
    fig = plt.figure(figsize=(12, 8))       
    ax1 = fig.add_subplot(111)

    # Scatter plots for each model
    ax1.scatter(image_names[1:650], processing_times_model_1_jittered[1:650], color='green', s=30, alpha=0.5, label="RCNN+FFNN")
    ax1.scatter(image_names[1:650], processing_times_model_2_jittered[1:650], color='red', s=30, alpha=0.5, label="YOLO+FFNN")
    ax1.scatter(image_names[1:650], processing_times_model_3_jittered[1:650], color='orange', s=30, alpha=0.5, label="RCNN+LSTM")
    ax1.scatter(image_names[1:650], processing_times_model_4_jittered[1:650], color='blue', s=30, alpha=0.5, label="YOLO+LSTM")
    
    # Add titles and labels
    plt.title('Model Inference Time Comparison (Logarithmic Scale with Jitter)')
    plt.xlabel('Image Index')
    plt.ylabel('Processing Time (ms)')
    plt.xticks(rotation=45, ha='right')

    # Set a logarithmic scale for the y-axis
    ax1.set_yscale('log')

    # Adjust y-axis limits to zoom in on the area of interest
    ax1.set_ylim([10, 3000])  # Adjust based on your data range

    # Add a larger legend to differentiate between models
    plt.legend(fontsize=12)

    # Add a grid
    ax1.grid(True, linestyle='--', color='gray', linewidth=0.5)

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.savefig('RCNN_VS_YOLO+FFNN+LSTM.jpg', dpi=300)
    plt.show()
    
else:
    print("Lengths of the datasets do not match. Cannot plot!")
```
