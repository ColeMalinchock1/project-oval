# Faster R-CNN Localization

Two Procedure: 
1. Faster R-CNN object detector (Faster Region Based Convolutional Neural Network)
   Output of RCNN:
   * Bounding Boxes: The coordinates and size of the detected objects.
   * Labels: The class name of the detected objects.
   * Scores: The confidence scores indicating the likelihood of the objects being correctly classified.
2. FFNN (Feedforward Neural Network)
   Output of FFNN:
   * Geolocation data(lat, lon) and orientation.

   

# Important points for Faster R-CNN for Landmark Detection
Comprises of two modules 
1. Fast R-CNN Object Detector:
   * Convolutional and Max Pooling Layers: Process entire map to generate a convolutional feature map(Lower-dimensional Representation of the input image, capturing essential features).
   * Region of Interest (RoI) Pooling Layer: Extracts fixed length feature vectors from the feature map for each proposed region.
   * Fully Connected Layer: Determines the class of the object, gets the corrdinates of the bounding boxes to more accurately fit the detected objects
2. Region Proposal Network(RPN):
   * Helps guide the Fast R-CNN detector by focusing on specific locations within the image where objects are likely to be found.

# Structure of the dataset
Each set of data in the dataset includes:

1. An image.
2. Latitude angle.
3. Longitude angle.
4. Magnetic compass orientation.
5. Bounding boxes of landmarks in the image.
6. Labels of landmarks for the bounding boxes.
