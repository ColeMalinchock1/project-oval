```python
# Preprocessing of the image to the require size which is 320, 240
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
read_img = cv2.imread("test_img1.jpg")
print(f"Size of image {read_img.shape}")
print(f" Height {read_img.shape[0]}")
print(f" Width {read_img.shape[1]}")
```

```python
# Converting the img to size (320, 240)
resize_img = cv2.resize(read_img, (320, 240))/255.0
print(f"Size of image {resize_img.shape}")
print(f" Height {resize_img.shape[0]}")
print(f" Width {resize_img.shape[1]}")
```

```python
# Resized Image 
resize_img_rgb = cv2.cvtColor((resize_img * 255).astype('uint8'), cv2.COLOR_BGR2RGB)
plt.imshow(resize_img_rgb)
plt.title('Resized Image')
```

```python
# Initial Image
read_img_rgb = cv2.cvtColor(read_img, cv2.COLOR_BGR2RGB)
plt.imshow(read_img_rgb)
plt.title("Initial Size")
```
